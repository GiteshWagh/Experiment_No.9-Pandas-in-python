```python
# Basics of Pandas
# Pandas : Panel Data

import pandas as pd # Importing pandas
series = pd.Series([11,22,33,44,55]) # Creating series
print(series) # Printing the created series

series1 = pd.Series([11,22,33,44,"55"])
print(series1)

# Creating DataFrame
data = {
    "Name" : ["a","b","c","d"],
    "Age" : ["11","22","33","44"],
    "Marks" : [44,55,66,77]
}

df = pd.DataFrame(data)
print(df)

print(df.head()) # Accessing first 5 rows
print(df.tail()) # Accessing last 5 rows

print(df.shape) # Gives the shape of the dataframe()

print(df.ndim) # Gives the dimensions of the dataframe

print(df.size) # Gives the size of the data frame (number of elements)

print(df.columns) # Gives the names of the columns

print(df.dtypes) # Gives the data types of the columns

# For accessing individual column
print(df["Name"]) # Printing the data of column Name
print(df["Age"])  # Printing the data of column Age
print(df["Marks"])  # Printing the data of column Marks

# For acessing specific rows
print(df.loc[0]) # Accessing first row
print(df.loc[1]) # Accessing second row
print(df.loc[2]) # Accessing third row

# For Adding New Column
df["Grade"] = ["A+","B+","C","Fail"]
print(df)

# For Updating The Data In Dataframe
df.loc[0,"Name"] = "Gitesh"
df.loc[0,"Age"] = "18"

print("\n",df)

# For Deleting The Column
df.drop("Grade",axis=1,inplace=True)
print("\n",df)

print(df["Marks"].mean()) # Mean of th perticular column data
print(df["Marks"].median()) # Median of th perticular column data
print(df["Marks"].min()) # Minimum value of th perticular column data
print(df["Marks"].max()) # Maximum value of th perticular column data

# Applying the conditions
df["Result"] = ["Pass" if marks >= 40 else "Fail" for marks in df["Marks"]] # Adding new columns with conditions applied
print(df)
```
